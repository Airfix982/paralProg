```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./csv/performance.csv')
dataOMP = pd.read_csv('./csv/performance_openMP.csv')
print(dataOMP)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['MatrixSize'], data[' AverageDuration'], marker='o')
plt.title('Average Duration vs Matrix Size')
plt.xlabel('Matrix Size')
plt.ylabel('Average Duration (ms)')
plt.grid(True)
plt.savefig('./png/avg_dur_not_paral.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='MatrixSize', y=' AverageOperations', marker='o', color='orange')
plt.title('Average Number of Operations for Matrix Multiplication')
plt.xlabel('Matrix Size')
plt.ylabel('Average Operations')
plt.grid(True)
plt.savefig('./png/avg_oper_not_paral.png')
plt.show()
```
